```python
code_list = list(set(df['stock_code'].values))
code_list.sort()
```

```python
resample_df = pd.DataFrame()
for code in code_list:
    rdf = df[df['stock_code'] == code].resample(rule = '1M', 
                                        on = 'trade_date', base = 0, 
                                        label='right', closed='right').agg({'trade' : 'last', 
                                                            'volume' : 'sum', 'amount' : 'sum', 
                                                            'low' : 'min', 'high' : 'max', 'code' : 'count'})
    rdf.reset_index(inplace = True)
    rdf.insert(0, column = 'stock_code', value = code)
    resample_df = resample_df.append(rdf)
```
